# B. Number of persons state wise vaccinated for first dose in India 


```python
covid_df2
```

```python
df1=covid_df2[covid_df2.State!='India']
df1.head()
```

```python
df_sorted = df1.sort_values('First Dose Administered', ascending=True)
```

```python
df = df_sorted[['State', 'First Dose Administered', 'Second Dose Administered']]
```

```python
grp_data=df.groupby("State").sum()
grp_data
```

```python
grp_column = grp_data['First Dose Administered']
grp_column 
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15,8))

sns.barplot(x=grp_data.index, y="First Dose Administered", data=grp_data, color="blue")
plt.xticks(rotation=90)
plt.title("Total First Dose Administered by State")
plt.xlabel("State")
plt.ylabel("Total First Dose Administered")
plt.show()
```

```python
grp_column = grp_data['Second Dose Administered']
grp_column 
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15,8))

sns.barplot(x=grp_data.index, y="Second Dose Administered", data=grp_data, color="green")
plt.xticks(rotation=90)
plt.title("Total Second Dose Administered by State")
plt.xlabel("State")
plt.ylabel("Total Second Dose Administered")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the data on the axis
ax.plot(covid_df3['State'], covid_df3['TotalSamples'], label='Total Samples')
ax.plot(covid_df3['State'], covid_df3['Positive'], label='Positive')
ax.plot(covid_df3['State'], covid_df3['Negative'], label='Negative')
ax.set_ylim(0, 10000000)
# Set the title, axis labels, and legend
ax.set_title('COVID-19 Testing in India')
ax.set_xlabel('State')
ax.set_ylabel('Number of Samples')
ax.legend()
plt.xticks(rotation=90)
plt.yticks(rotation=90)
plt.show()
```
