# B. Number of persons state wise vaccinated for first dose in India 


```python
df_sorted = covid_df2.sort_values('First Dose Administered', ascending=True)
```

```python
df = df_sorted[['State', 'First Dose Administered', 'Second Dose Administered']]
```

```python
grp_data=df.groupby("State").sum()
grp_data
```

```python
grp_column = grp_data['First Dose Administered']
grp_column 
```

```python
grp_column = grp_data['Second Dose Administered']
grp_column 
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15,8))

sns.barplot(x=grp_data.index, y="First Dose Administered", data=grp_data, color="blue")
plt.xticks(rotation=90)
plt.title("Total First Dose Administered by State")
plt.xlabel("State")
plt.ylabel("Total First Dose Administered")
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(15,8))

sns.barplot(x=grp_data.index, y="Second Dose Administered", data=grp_data, color="green")
plt.xticks(rotation=90)
plt.title("Total Second Dose Administered by State")
plt.xlabel("State")
plt.ylabel("Total Second Dose Administered")
plt.show()
```
